```python
from matrixpack import target_pattern, generate_mask
from matrixpack.constants import dx, dz, MagAM, fSLM, SLMpixel, SLMheight, SLMwidth, beamD
from LightPipes import LPtest, nm, um, mm, Intensity
import matplotlib
%matplotlib inline
import pylab as plt
import sys
import os
```

```python
print('Environment: {}'.format(sys.exec_prefix))
print('Executable: {}'.format(sys.executable))
print('Python version: {}'.format(sys.version))
print('Matplotlib', matplotlib.__version__)
LPtest()
```

```python
cwd = os.getcwd()
#print('cwd: {}'.format(cwd))
outPath = cwd+'\masks'
```

### Computation
---
Convention for `movebeams_um`

        movebeams_um=[beam1,beam2,beam3,
                      beam4,beam5,beam6,
                      beam7,beam8,beam9]
        where
        far = beams1-4-7
        mid = beams2-5-8
        near = beams3-6-9  
    +ve values for `movebeams_um` moves the beam UP in sample space i.e., along +Z in the microscope
    
    1. Define target pattern `target_pattern()`
    2. Compute phase mask to generate the above target pattern when applied to an SLM `generate_mask`

```python
wavelength = 488*nm
Niter = 100
tag='MATRIX'
movebeams_um=[0,0,0,
              0,0,0,
              0,0,0]

[SLMfield, target, footer] = target_pattern(wavelength,beamD,
                                    fSLM,SLMpixel,SLMwidth,SLMheight,
                                    dx,dz,MagAM,
                                    tag,movebeams_um)

generate_mask(wavelength, SLMfield, target, 
              SLMwidth, SLMheight, 
              outPath, Niter,
              footer, tag)
```

```python
wavelength = 488*nm
Niter = 100
tag='MATRIX'
movebeams_um=[3,3,3,
              0,0,0,
              -3,-3,-3]

[SLMfield, target, footer] = target_pattern(wavelength,beamD,
                                    fSLM,SLMpixel,SLMwidth,SLMheight,
                                    dx,dz,MagAM,
                                    tag,movebeams_um)

generate_mask(wavelength, SLMfield, target, 
              SLMwidth, SLMheight, 
              outPath, Niter,
              footer, tag)
```

```python
wavelength = 488*nm
Niter = 100
tag='MATRIX'
movebeams_um=[-3,-3,-3,
              0,0,0,
              3,3,3]

[SLMfield, target, footer] = target_pattern(wavelength,beamD,
                                    fSLM,SLMpixel,SLMwidth,SLMheight,
                                    dx,dz,MagAM,
                                    tag,movebeams_um)

generate_mask(wavelength, SLMfield, target, 
              SLMwidth, SLMheight, 
              outPath, Niter,
              footer, tag)
```

### Plot Field Intensities
---

```python
#Plot field intensities
ISLM = Intensity(SLMfield)
high= 1000
hoff= 100
wide= 600
woff= 300
top= len(target)//2 - high//2 + hoff
bottom= len(target)//2 + high//2 - hoff
left= len(target)//2 - wide//2 
right= len(target)//2 + wide//2 - woff

plt.figure(figsize=(17,17))
plt.subplot(1,2,1)
plt.title('Intensity (at SLM)')
plt.imshow(ISLM,cmap='jet')
plt.subplot(1,2,2)
plt.title('Intensity (at target)')
plt.imshow(target[top:bottom,left:right], cmap='gray')
plt.show()
```
